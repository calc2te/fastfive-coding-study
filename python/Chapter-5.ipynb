### 히스토그램 그려보기

```python
import csv

array = []

f = open('./ta_20200701230055.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
for row in data:
    array.append(row)
f.close()
```

```python
array = array[8:]
array = array[:-1]
```

```python
avg = []

for a in array:
    avg.append(float(a[2])) # float를 이용해서 문자를 실수로 바꿉니다.
```

- 기존 그래프

```python
import matplotlib.pyplot as plt

plt.plot(avg)
plt.show()
```

- plot을 hist로 바꿔서 히스토그램으로 나타내보기

```python
plt.hist(avg)
plt.show()
```

- 히스토그램의 느낌적인 느낌을 보면...

```python
plt.hist([1, 1, 1, 2, 2, 3, 4, 4, 4, 4])
plt.show()
```
